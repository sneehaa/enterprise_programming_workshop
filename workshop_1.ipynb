```python
%%writefile cudahelloworld.cu
#include <stdio.h>
#include <cuda.h>
#include <cuda_runtime.h>

__global__ void kernel()
{
  printf("Hello world from kernel\n");
}
int main()
{
  printf("Hello world from host\n");
  kernel<<<3,1>>>();

  cudaDeviceSynchronize();

  return 0;
}
```
